This script plots temperature and THI in the same graph for one farm

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 200 #Larger figures
import matplotlib.ticker as ticker
sns.set()
```

```python
farmNames = ["ad0a39f5", "752efd72", "5c06d92d", "540275a1", "a624fb9a", "f454e660", "ab18b151"]

rDict = {"a624fb9a": "Farm 1",
    "ab18b151": "Farm 2",
    "540275a1": "Farm 3",
    "5b581702": "Farm 4",
    "5c06d92d": "Farm 5",
    "a756bc39": "Farm 6",
    "ad0a39f5": "Farm 7",
    "752efd72": "Farm 8",
    "f454e660": "Farm 9"}
```

```python
for name in farmNames:
    sName = rDict[name]
    fName = f"{name}_weather_daily.csv"
    df = pd.read_csv(fName)
    df["Timestamp"] = pd.to_datetime(df["Timestamp"])
    df = df.set_index("Timestamp")


    fig = plt.figure()
    ax = sns.lineplot(df[["THI_mean", "THI_adj_mean"]], linewidth = 0.7, alpha = 0.7)
    ax.legend(['First line', 'Second line'])
    ax2 = ax.twinx()
    sns.lineplot(df["T_C_mean"], ax=ax2, color = "green", legend='brief', label="Temperature", linewidth = 0.7, alpha = 0.7)
    ax2.grid(None)
    ax.set_ylabel("THI")
    ax2.set_ylabel("Temperature (C)")
    ax.set_xlabel("")
    ax.legend(loc=0)
    ax2.legend(loc=2)
    fig.autofmt_xdate(rotation=45)
    plt.title(f"THI and temperature for {sName}", size = 16)
    plt.show()

```
