This script plots temperature and THI in the same graph for one farm

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 200 #Larger figures
import matplotlib.ticker as ticker
sns.set()
```

```python
df = pd.read_csv("ad0a39f5_weather_daily.csv", usecols=["Timestamp", 'T_C_mean', 'THI_mean', 'THI_adj_mean'])
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
df = df.set_index("Timestamp")
```

```python
colors = ["#CE0037", "#00b0b9", "#FFB81C", "#509E2F", "#672146", "#007681", "#996017", "#79863C"]
customPalette = sns.set_palette(sns.color_palette(colors))
```

```python
fig = plt.figure()
ax = sns.lineplot(df[["THI_mean", "THI_adj_mean"]], linewidth = 0.7, alpha = 0.7)
ax.legend(['First line', 'Second line'])
ax2 = ax.twinx()
sns.lineplot(df["T_C_mean"], ax=ax2, color = "green", legend='brief', label="Temperature", linewidth = 0.7, alpha = 0.7)
ax2.grid(None)
ax.set_ylabel("THI")
ax2.set_ylabel("Temperature (°C)")
ax.set_xlabel("")
ax.legend(loc=0)
ax2.legend(loc=2)
fig.autofmt_xdate(rotation=45)
plt.title("THI and temperature for Farm 7", size = 16)
plt.show()
```
