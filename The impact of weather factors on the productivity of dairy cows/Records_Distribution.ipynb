Explore the distribution of records for each farm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib as mpl
import seaborn as sns
sns.set()
```

```python
milk = pd.read_csv("AllMilkings.csv", usecols=["StartDate", "FarmName_Pseudo", "SE_Number", "SE_Number", "LactationNumber", "DaysInMilk", "TotalYield"])
rDict = {"a624fb9a": "Farm 1",
        "ab18b151": "Farm 2",
        "540275a1": "Farm 3",
        "5b581702": "Farm 4",
        "5c06d92d": "Farm 5",
        "a756bc39": "Farm 6",
        "ad0a39f5": "Farm 7",
        "752efd72": "Farm 8",
        "f454e660": "Farm 9"}
milk = milk.replace({"FarmName_Pseudo": rDict})
farms = ['Farm 1', 'Farm 2', 'Farm 3', 'Farm 4', 'Farm 5', 'Farm 6', 'Farm 7', 'Farm 8', 'Farm 9']
for farm in farms:
    day_yeild = milk[milk["FarmName_Pseudo"] == farm]
    fig, ax = plt.subplots(1,1)
    sns.scatterplot(x = "StartDate", y = "TotalYield", data = day_yeild, ax = ax)
    tick_spacing = 180
    ax.xaxis.set_major_locator(mpl.ticker.MultipleLocator(tick_spacing))
    ax.set_title(f"Distribution of records for {farm}",x = 0.5, y = 1, fontsize = 16)
    plt.show()
```

```python
# Plot the distribution of records for Farm 4 and Farm 6
farms = ["Farm 4", "Farm 6"]
fig, axs = plt.subplots(1,2)
fig.tight_layout(pad = 2.5)
axs = axs.ravel()
for i in range(2):
    day_yeild = milk[milk["FarmName_Pseudo"]==farms[i]]
    sns.scatterplot(x = "StartDate", y = "TotalYield", data = day_yeild, ax = axs[i])
    tick_spacing = 90
    axs[i].xaxis.set_major_locator(mpl.ticker.MultipleLocator(tick_spacing))
    axs[i].set_title(f"{farms[i]}")
plt.suptitle("Distribution of records for Farm 4 and Farm 6",x = 0.5, y = 1, fontsize = 16)    
plt.savefig("distribution", bbox_inches = 'tight')
plt.show()
```

With the records for only several days, Farm 4 and Farm 6 are not considered in the further analysis.
